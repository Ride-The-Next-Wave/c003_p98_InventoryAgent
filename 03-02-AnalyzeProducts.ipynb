# 3.2 Analyze Product Data
This notebook analyzes the cleaned product data in preparation for modeling. The ultimate goal is to match cleaned articles to the products based on their cleaned product data.

```python
import pandas as pd
```

```python
import json
import re
import matplotlib.pyplot as plt
from collections import Counter

# Load product data
with open("data/product-data-cleaned.json") as f:
    product_data = json.load(f)

# Define basic English stopwords + your custom ones
manual_stopwords = {
    "the", "and", "a", "to", "of", "in", "is", "it", "you", "that", "this", "on", "for", "with", "as",
    "was", "but", "are", "they", "at", "be", "from", "by", "an", "or", "not", "have", "we", "has",
    "q", "ti", "one", "get", "right", "full", "go", "can", "do", "i", "our", "your", "all", "out", "what", "want"
}

# Tokenize and filter stopwords
product_tokens = [
    [token for token in re.findall(r'\b\w+\b', item["clean_text"].lower()) if token not in manual_stopwords]
    for item in product_data
]


# Flatten all tokens
all_tokens = [token for tokens in product_tokens for token in tokens]

# Count token frequency
counter = Counter(all_tokens)
most_common = counter.most_common(20)

# Plot most frequent words
words, counts = zip(*most_common)
plt.figure(figsize=(12, 5))
plt.bar(words, counts)
plt.xticks(rotation=45)
plt.title("Top 20 Most Frequent Words in Product Descriptions")
plt.show()


# # Preview some example token lists
# for i in range(3):
#     print(f"Product {i} tokens:", product_tokens[i][:20])
```
